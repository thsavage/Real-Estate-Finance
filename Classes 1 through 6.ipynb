## Basic accounting

* Let $x$ denote the radial distance from an employee's residence to the CBD. 
* $x_1 > x_2 \Rightarrow$ it is more costly for employee$_1$ to travel to work than employee$_2$.
    * Two components: price (direct cost) and time (opportunity cost)
* Let $t$ denote the per-distance cost of commuting, which implies that $x \cdot t$ is the total cost of a one-way journey.  In the U.S., this is measured as mile-dollars.  In much of Europe, this is meter-euros or meter-pounds.
    * $t$ also captures alternative transportation modes.  One journey on the NYC ferry system is the same as Midtown Center to Washington Square.
    * Tim to Schack on the subway: $0.10 \cdot 2.75 = 0.275$ mile-dollars .
    * Tim to Boston on the Acela: $4 \cdot 200 = 800$ mile-dollars.
* Let $y$ denote wage income from employment in the CBD.  Disposible income $i$ is, therefore, $y-x \cdot t$.
    * The expression in increasing in $y$ but decreasing in both $x$ and $t$, and the effects are linear.
    * The longer the commute distance, the lower is disposible income.  
* Consumption consists of two goods: transportation and bread. 
    * Bread consumption is denoted by $c$, the price of which is normalized to 1.  
    * Housing consumption is denoted by $q$, but the physical units corresponding to $q$ must be chosen. 
        * Housing is complicated and varies considerably by its characteristics.  Let $p$ be a simple characteristic: rental price per square unit. 
        * Housing costs are, therefore, $q \cdot p$.
* Consumer budget constraint is:
    * $c + q \cdot p = yx \cdot t$.
    * Consumption is on the left of the equal sign.
    * Income is on the right of the equal sign.
    * There are no savings, so all income is consumed.
* The goal of the employee is maximize her utility over $c$ and $q$ given the budget constraint.

## An introduction to Pareto efficiency and sources of market failure

* **Class Discussion**: What does efficiency mean to you?


* Economists have established a baseline definition of **efficiency**, originally articulated by economist Wilfredo Pareto.  
    * He also articulated the so-called "80-20 rule" with regard to wealth distributions.  80% of the wealth is owed by 20% of a population.  He is also created with a probability distribution, called the **Pareto distribution** to describe the distributions of income and wealth.
    * The 80-20 rule is a power law consistent with the fractal mathematics of [Benoit Mandelbrot](https://en.wikipedia.org/wiki/Benoit_Mandelbrot).
* It essentially describes the redistribution of resources (good and services) in an economy.
* If redistribution can make at least one person better off **without making anyone else worse off**, then such a redistribution is said to be Pareto enhancing or Pareto improving.
* This process is repeated until at one person must be made worse off to make at least one person better off.
    * At that point, the equilibrium is said to be **Pareto efficient**.
    * The set of all points at which this is true is called a **Pareto frontier**.
    * This is not a condition of fairness, but of efficiency.

* A market failure is a situation in which the allocation of resources (good and services) is inefficient.
    * As a result, some action by an external actor to the market can increase welfare through the re-allocation of resources.
    * If the Pareto inefficient outcome arises in a market without outside intervention, then there is a market failure because at least one market participants could be made better off without making other market participants worse off.
    * The correction of the market failure improves aggregate welfare.

## Other distortions affecting urban land use

* Local laws and regulations can generate distortions in "optimal" land use.
    * These are not market failures, but can be summarized by the results of local pressure best summarized by the phrase **NIMBYism**: "Not In My Back Yard"-ism.
    * One arises because of the need to raise revenue to pay for local infrastruture such as sewers or to fund local public education.
        * For example, extending sewer lines into less densely populated suburbs may be more costly on the margin.
        * Higher expenses are distributed across all citizens, urban and suburban.
        * Higher expenses may result in higher tax rates across all citizens, with current residents bearing a disproportionate cost relevant to "new comers".  
            * This constitutes a form of tax subsidy for the new suburban residents.
        * In response, local governments in the U.S. have begun to impose user (or impact) fees on new residents.  Such fees might also fall on developers when public roads are extended.
    * In the U.S., however, the larger distortion arises from federal government subsidies of home ownership by allowing the tax deductibility of mortgage interest.
        * Such subsidies may not drive homeownership, but rather drive larger dwellings per person in equilibrium.  Large houses consume more space and tend to be located in suburbs.
        * When the U.K. eliminated its mortgage interest subsidies under Thatcher, there was no change in homeownership rates.
        * Taxes distort, **as do subsidies**.
    * In the U.S., gasoline taxes are very low relative to Europe and parts of Asia.
        * This is an implicit subsidy to car ownership and longer and/or more frequent journeys in equilibrium.
    * In the U.S. and, **especially in Europe**, large subsidies are paid to farmers that consistute a very small fraction of the population or economic output.
        * Such subsidies **distort land use** and, in equilibrium, place a floor on the price of agricultural land

## Rent control and its potential impact on development

* Rent control and regulations exist in many U.S. cities, NYC being one of the most famous examples.
    * **Origin**: wage and price controls imposed in the U.S. during WWII.
        * These controls essentially created employer-based health insurance in the U.S. as it was not covered (largely because health care involved the use of penicillin).
        * They also gave rise to regulations governing rents landlords are able to charge tenants in MF.
* In their purest form, a rent control is similar to a minimum wage (acting as a ceiling on rents rather than a floor on wages).
    * It establishes a ceiling on rent levels.
    * If said ceiling is below an equilibrium rent level, it creates artificial scarsity in equilibrium.
    * If said ceiling is not sufficient for owners to break even on an NOI basis, investment will not be sufficient to maintain properties.  As a result, *housing quality* will decline over time.
    * This is the insidious nature of rent control.  Loss of quality is gradual, while the political pressure to enact is immediate, especially in an election cycle in an expensive city like NYC.
    * Moreover, benefits accrue to *current* tenants, while *future* tenants are receiving lower quality housing over time.
        * Gyourko and Linneman (1990) examine data from NYC and find that rent-controlled buildings are more likely to be *dilapidated* than buildings without rent control.
        * Diamond *et al.* (2019) use "natural experiment" in San Francisco and find that rent control reduces both mobility and housing supply.
* In a weaker form, rent control is called **rent stabilization**.
    * It controls the rate of growth in rent from some base level, which itself maybe arbitrary.
    * Many new MF units in NYC are rent stabilized initially and only for a limited period of time after which landlords may charge market rents as well increases that the market will bear.
* Compare and contrast: rent control and a minimum wage.
    * Rent control is an artificial ceiling on rents.  A minimum wage is an artificial floor on wages.
        * Empirical studies have demonstrated the adverse impacts of rent control on the quality of existing builings.  Empirical studies have not demonstrated the adverse impacts of minimum wages.  
        * Why?
        * One argument identifies the nature of housing markets versus labor markets.
        * In labor markets, employers may have market power which results in equilibrium wages that are lower than would prevail in absence of the employer market power.
        * In rental markets, for this argument to apply, landlords would have market power, say do to scarcity.  The problem, however, is that, if true, the rent control ceiling *exacerbates this lack of supply* (or its quality).  As a result rent control *increases* landlord market power.  
* Rent control and mobility.
    * Rent control and public housing are often presented as temporary policies while new stock is built or existing stock is improved.
    * Arguably, rent control reduces mobility because renters are paying below-market rents, and moving might void this subsidy.
    * It is also true that public housing has many generations of the same family.
    * Hence, neither policy has proved to be *temporary*.

## Impact on development

* Rent control creates an artificial ceiling on rents.
* Consider a developer who is focused on new MF development or the refurbishment of an existing building.
    * We have discussed the use of DCF models to calculate IRRs.
    * The imposition of the rent ceiling reduces NOI and therefore the IRR **after the fact**.
    * As a result, rent control negates the profit from construction or redevelopment.
* We can explore this idea using a simple *stock-flow* model.
    * Stock is existing housing.
    * Flow is new construction or redevelopment.
    * Assume for the sake of argument that all housing is rented rather than owner-occupied.
    * **Graph 7**.
        * The left panel is the stock, while the right panel is the flow.
        * The left panel has an elastic demand curve but an inelastic supply curve that together determine the equilibrium rental price, $p$.  For example, $p_0$ is equilibrium rental price for housing stock, $H_0$. 
        * Why inelastic?  Landlords do not want any vacancy.
        * The right panel visualizes flow stock, $S_{\Delta h}$.  
        * Key reflection in the two graphs: When $p$ is high, flows are high.  As $p$ falls, so do the flows.  At some price, $p_E$ flow stock is zero, and the market is in equilibrium.

        * If a rent control law imposes a rental price below $p_E$, housing stock will decline.

## Rent control and consensus opinion

* The broad consensus among economists is that rent control and stabilization policies, while perhaps driven by good intention, are examples of policies with adverse unintended consequences.
    * This is not true of minimum wage laws.
* They allow local politicians to redirect citizens' ire away from policies that constrain supply toward the development industry.
    * This perversely provides the development community more market power, even if the media headlines are bad.
* They allow local politicians to avoid much more important issues regarding the delivery of local public services, such as local school and road quality.
* High rents in urban locations such as NYC are exacerbated by artificial supply constraints in the face of strong demand.

### Pollution and crime: negative externalities in the urban built space
### Application of pigouvian taxation

#### Digression on microeconomics
* A **marginal benefit** curve is a measure of the incremental benefit of the increase in *something*.
* A **marginal cost** curve is a measure of the incremental cost of the increase in *something*.
* How to determine the optimal level of *something*?  Apply the following rule:
    * Increase the level of *something* as long as the marginal benefit outweighs the marginal cost.
    * The difference between the two is the marginal gain (or loss) of *something*.
    * For firms in a competitive market, this rule gives rise to "equating marginal revenue with marginal cost".
    * For individuals who take prices as given, this rule gives rise to "equating the ratio of marginal utility to the ratio of prices."
    * In finance, it also gives rise to the Capital Asset Pricing Model per Lucas (1978).

### Basic framework

* Consider three houses and a single polluting source.
* Pollution creates small particulate matter that can harm health, as well as adversely impact the value of housing.
    * Large literature to measure the magnitude: See Chay and Greenstone (2005) and its update, Berkman, Hubbard and Savage (2011).
    * Adverse impacts are not small $\Rightarrow$ enforcement benefits are large.  Tougher enforcement of U.S. air-quality standards over the 1970s generated an aggregate property-value increase of 45 billion.
   
* **Graphs 8 and 9**
    * Let $D(p)$ be a measure of damage done by pollution, $p$.
    * Let $D^\prime(p)$ be positive in $p$.
    * Pollution is harmful the for residents, but necessary (and therefore beneficial) for the firm.
        * $B(p)$ and $B^\prime(p)$ measure the benefit and marginal (or incremental) benefit to the firm.
        * At $\overline{p}$, no pollution abatement occurs.  Small increases in abatement are inexpensive for the firm.
    * There is no marketplace for pollution abatement.
    
* Basic Coasian principle: residents could pay firm not to pollute or firm could pay residents to compensate them for the pollution.  Either approach is, from economics, indistinguisable.

### Socially-optimal pollution

* It is not zero, but because there is no market with transparent prices, the amount of abatement is not socially optimal.  
* In absence of controls, the firm would pollute up to the point that the marginal benefit from doing so is zero or $B^\prime(p) = 0$.
    * The marginal benefit and damage curves allow us to expore the socially-optimal level of pollution.
* **Graph 10**
    * Apply the rule from above.
    * Start at no pollution, $p=0$, and increase the level of pollution as long as the marginal benefit exceeds the marginal damage.  
    * Stop when they are equal.

#### Getting to $p^*$
* Coasian negotiation between firm and residents.
    * Transactions cost must be low, and no party should have bargaining power.
    * Property rights must be assigned: someone must "own the air".
        * Firm has the right to pollute, and residents pay the firm not to pollute.  (Good luck with that.)
        * Residents have the right to clear air.  Residents through government are compensated for losses by the firm.  
* Government mandates, such as the Clear Air Act in the U.S.
* Government uses Pigouvian tax.  [A.C. Pigou](https://en.wikipedia.org/wiki/Arthur_Cecil_Pigou).  The tax forces the firm to internalize the costs of pollution as captured by the marginal damage curve.  This is the basis for carbon taxes.  
    * Government creates a market that **could not** form on its own.  This was called "Cap and Trade" in the U.S.
        * Establish an overall cap on the level of carbon and create tradeable permits.  The least efficient firms will pay for the permits, and the most efficient will sell the permits.  This forces the least efficient firms to internalize the costs of pollution.
            * This was the basis for the [1990 Clear Air Act](https://environmentallaw.uslegal.com/federal-laws/clean-air-act/).
    * Auction markets for EM spectrum have been incredibly successful.
    * Auction markets for energy have been much less successful, at least in the U.S.
